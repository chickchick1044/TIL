```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="C:\Windows\Fonts\malgun.ttf").get_name()
rc('font', family=font_name)


df =pd.read_csv('credit_card_data.csv',encoding='utf8')
```

```python
df['region']="지역"
```

```python
df.head()
```

```python
sudo=['인천','경기']
chungnam=['대전','충남','충북']
gyungsang=['대구','울산','경북','경남']
gunra=['광주','전북','전남']
'''
"서울"
"부산"
"제주"
"강원"
'''
```

```python
for i in range(0,len(df['region'])):
    
    if df['city'][i] in sudo:
        
        df['region'][i]="수도권"
        
    elif df['city'][i] in chungnam:
        
        df['region'][i]="충청도"
        
    elif df['city'][i] in gyungsang:
        
        df['region'][i]="경상도"
        
    elif df['city'][i] in gunra:
        
        df['region'][i]="전라"
        
    elif df['city'][i]== '서울':
        
        df['region'][i]= "서울"
        
    elif df['city'][i]== '부산':
        
        df['region'][i]= "부산"
        
        
    elif df['city'][i]== '제주':
        
        df['region'][i]= "제주"
        
        
    elif df['city'][i]== '강원':
        
        df['region'][i]= "강원"
    
    else :
        df['region'][i] = "성별"
```

```python
총 대출약정금액과
총 대출금액,
은행업종 총 대출금액의 지표를 보면
서울지역에 80,90대의 총금액 및 평균이 다른지역과 평균에 비해
차별적으로 높음을 볼 수 있다. 

카드 업종 총 대출금액을 보면
수도권(서울+인경기)지역만 30대 대출보다 40대의 대출금액이
높음을 알 수 있다. 전체적인 양을 보더라도
30대의 청년보단 40,50대의 대출액이 높음을 알 수 있다. 

보험업종 총 대출금액을 확인해 보면 제주의 지역에서
월등하게 80대의 총액의 합이 큼을 알 수 있다.
이는 데이터의 오류 혹은 제주도의 특징일 수 있다. 

저축은행업종의 경우 모든 부문에 있어 
10대의 저축은행  대출이 많음을 알 수 있다. 
연령이 낮을 수록 저축은행 대출금액이 높음을 확인할 수 있다.

한도대출의 경우 
서울, 수도권, 부산의 경우 30 대이후로는 동일 수준의 금액으로 보이지만
타 지역의 경우는 연령대 별로 차이가 있음이 나타난다.

신용대출 금액 총합의 경우 *금융회사가 고객의 신용도를 판단하여 해주는 대출이 신용대출이다. 
서울 90대의 월등한 금액의 차이가 있음이 나타난다.

담보대출 또한 *담보대출은 신용대출보다 장기간이고 금액이 크며 대출금리가 낮은 것이 일반적이다.
전체적으로는 서울/수도권의 평균 대출금액이 높음.
서울 70~90대의 월등한 금액의 차이가 있음이 나타난다.


신용카드 일시불과 할부 금액을 비교하면
일시불의 경우 30대가 40,50대보다 높은 반면
할부의 경우 40,50대가 30대보다 높은 경향을 보인다.
```
