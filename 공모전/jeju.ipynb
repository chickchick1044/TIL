```python
df = pd.read_csv('jeju_financial_life_data.csv')
```

```python
df.head()
```

```python
df['age_sep'] = df['age'].apply(lambda x:str(x)[0] + str(0) + str("대"))
```

```python
df['age_5sep']=1
```

```python
# 5세구간

for i in range(0,len(df['age_5sep'])):
    
    if df['age'][i] == 24:        
        df['age_5sep'][i]="24세 이하"
        
    elif df['age'][i] == 29:
        df['age_5sep'][i]="25~29세"
        
    elif df['age'][i] == 34:  
        df['age_5sep'][i]="30~34세"

    elif df['age'][i] == 39:
        df['age_5sep'][i]="35~39세"
        
    elif df['age'][i] == 44:
        df['age_5sep'][i]="40~44세"
        
    elif df['age'][i] == 49:
        df['age_5sep'][i]="45~49세"
        
        
    elif df['age'][i] == 54:
        df['age_5sep'][i]="50~54세"
    
    elif df['age'][i] == 59:
        df['age_5sep'][i]="55~59세"
        
    elif df['age'][i] == 64:
        df['age_5sep'][i]="60~64세"
        
    elif df['age'][i] == 69:
        df['age_5sep'][i]="65~69세"
        
    elif df['age'][i] == 74:
        df['age_5sep'][i]="70~74세"
        
    elif df['age'][i] == 79:
        df['age_5sep'][i]="75~79세"
        
    elif df['age'][i] == 99:
        df['age_5sep'][i]="80세 이상"
```
