## 위의 그래프에서 남자는 은행, 할부 대출금액이 더 높다

여자는 보험, 카드 대출금액이 더 높다

저축은행은 10대 여자가 비정상적으로 높아서 데이터 검증 필요, 

20~60대 <- 돈을 버는 주체는 남자가 저축은행 대출비용이 더 높으나
 
 10대,70~90대 <- 돈을 벌지 않는 쪽에서는 여자의 저축은행 대출비용이 더 높게 나타남

```python
f, axes = plt.subplots(2,3, figsize = (20,10), sharex=True)
sns.stripplot(x = 'ages', y = 'monthly_cd_loan', hue='sex', data = df_no_city, ax=axes[1,1])
sns.boxplot(x = 'ages', y = 'monthly_bk_loan', hue='sex', data = df_no_city,ax=axes[0,1])
sns.boxplot(x = 'ages', y = 'monthly_installments_loan', hue='sex', data = df_no_city,ax=axes[0,2])
sns.boxplot(x = 'ages', y = 'monthly_insurance_loan', hue='sex', data = df_no_city,ax=axes[1,0])
sns.boxplot(x = 'ages', y = 'monthly_sbk_loan', hue='sex', data = df_no_city,ax=axes[1,2])
plt.show()
```
