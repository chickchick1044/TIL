Seaborn: matplotlib을 기반으로 다양한 색상 테마와 통계용 차트 등의 기능을 추가한 시각화 패키지

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

> 1차원 실수 분포 플롯: rugplot, kdeplot, distplot

```python
np.random.seed(0)
x = np.random.randn(100)
```

```python
# rugplot: 데이터 위치를 x축 위에 작은 선분(rug)로 나타내어\
# 실제 데이터들의 위치를 보여준다.
sns.rugplot(x)
```

```python
# kde: 커널밀도, 히스토그램보다 부드러운 형태의 분포 곡선을 보여주는 방법
sns.kdeplot(x)
```

```python
# distplot: 러그와 커널밀도 표시 기능이 있어서
# matplotlib의 hist보다 많이 사용된다.
sns.distplot(x, kde=True, rug=True)
```

> 2차원 실수 분포 플롯: jointplot

```python
# jointplot: 스캐터플롯
tips = sns.load_dataset("tips")
sns.jointplot(x="total_bill", y="tip", data=tips)
```

```python
iris = sns.load_dataset("iris")
sns.jointplot("sepal_width", "petal_length", data=iris, kind="kde", \
             space=0, color="g")
```

```python
sns.regplot(x="total_bill", y="tip", data=tips)
```

```python
sns.residplot(x="total_bill", y="tip", data=tips)
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips)
```

```python
sns.lmplot(x="total_bill", y="tip", col="smoker", data=tips)
```

```python
sns.barplot(x="day", y="total_bill", hue="sex", data=tips)
#막대의 높이는 평균, 에러바는 편차를 나타낸다.
```
