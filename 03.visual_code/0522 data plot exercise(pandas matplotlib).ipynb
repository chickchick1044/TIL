```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_url = 'data/housing.data'
df_data = pd.read_csv(data_url, sep='\s+', header=None)
df_data.columns = [
    'CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', \
    'PTRATIO', 'B', 'LSTAT', 'MEDV']
df_data.head()
```

```python
fig = plt.figure()
ax = []
for i in range(1,5):
    ax.append(fig.add_subplot(2,2,1))
    
ax[0].scatter(df_data["CRIM"], df_data["MEDV"])
ax[1].scatter(df_data["PTRATIO"], df_data["MEDV"])
ax[2].scatter(df_data["CRIM"], df_data["MEDV"])
ax[3].scatter(df_data["NOX"], df_data["MEDV"])
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(10.0, 10.0, forward=True)
ax = []
for i in range(1,5):
    ax.append(fig.add_subplot(2,2,i))
columns = ["CRIM", "PTRATIO", "AGE", "NOX"]
colors = ["b", "g", "c", "r"]
for i in range(4):
    ax[i].scatter(df_data[columns[i]], df_data["MEDV"],
    color=colors[i], label=columns[i])
    ax[i].legend()
    ax[i].set_title(columns[i])
plt.subplots_adjust(wspace=0, hspace=0)
plt.show()
```
