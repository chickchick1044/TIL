```python
import pandas as pd
import numpy as np
df = pd.DataFrame({
    'a':pd.date_range(start='2016-01-01', periods=20, freq='D'),
    'y':np.random.randn(20),
    'c':np.random.choice(['low', 'medium', 'high'], 20).tolist()
})
df2 = df.reindex(index=[0,2], columns=['a','c'])
df2
```

```python
df3 = pd.DataFrame(np.random.randn(10,2),
                  index=[1,4,6,2,3,5,9,8,0,7], columns=['col1', 'col2'])
# df3
df4 = df3.sort_index()
df4
df5 = df4.sort_index(ascending=False)
df5
```

```python
df5.sort_index(axis=1)
df5.sort_values(by='col1') # 값기준으로 변경, index 순서가 바뀜
df5.sort_values(by=['col1', 'col2']) # col1으로 정렬하고, 같은값이 있으면 col2기준으로 정렬
```

```python
# dataframe
# ['a'] -> column
# iloc[0] -> row select by index
# loc[0] -> row select by label
# df.iloc[0,0]
df5 = pd.DataFrame(np.random.randn(8,4))
df5
```
